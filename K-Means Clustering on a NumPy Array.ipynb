#### Section 1:  Exploratory Analysis

First we will import the 'matplotlib', 'numpy', & 'sklearn.cluster' packages.

```python
import matplotlib.pyplot as plt
import numpy as np
from sklearn.cluster import KMeans
%matplotlib inline
```

With the necessary packages imported, we can now prepare the data that is to be clustured. Here, we will be using a NumPy array with 10 rows & 2 columns consisting of randomized numbers.  

```python
X = np.array([[6,2], [9,16], [13,16], [20,8], [33,41], [88,69], [70,85], [61,79], [59,53], [82,94],])
```

Here, no preprocessing data is required because we are dealing with NumPy arrays. Let's draw a scatter plot to see if we can spot any clusters in advance before applying the algorithm.  

```python
plt.scatter(X[:,0], X[:,1], label = 'True Position')
plt.style.use('ggplot')
plt.xlabel('x values of the array')
plt.ylabel('y values of the array')
plt.title('Scatter Plot of the x vs y Values in our Array')
plt.show()
```

Note that the entry 'X[:,0]' selects the values in the x position of our array, & the entry 'X[:,1]' selects the values in the y position of our array. Without actually implementing the algorithm, it appears as though we may have two clusters. Namely, one cluster of 5 points in the top right, & another cluster of 5 points in the bottom left. There are possibly three clusters available. 

#### Section 2: Algorithm with Two Clusters 

We will now test if this is the case or not. Let's create two clusters. First let's create a 'KMeans' object & set the number of clusters to '2'.

```python
kmeans = KMeans(n_clusters=2)
```

Next we call the 'fit' method on 'kmeans' & pass the data ('X') that is to be clustered. 

```python
kmeans.fit(X)
```

The algorithm has generated the centroid values for the final clusters. Let's see what they are...

```python
print(kmeans.cluster_centers_)
```

Each row represents the coordinates for the centroids. In other words, the first centroid has the coordinates '[72.0, 76.0]' & the second centroid has coordinates '[16.2, 16.6]'. Now let's print the labels for the data point. 

```python
print(kmeans.labels_)
```

We have a one dimensional array of 10 elements corresponding to the clusters assigned to our 10 data points. This tells us that the first five points have been clustered together, & the last five points have been clustered together. The values '1' & '0' are used to represet cluster IDs. Now we can create another plot to see how the data has been clustered. 

```python
plt.scatter(X[:,0], X[:,1], c=kmeans.labels_, cmap='rainbow')
plt.style.use('ggplot')
plt.xlabel('x values of the array')
plt.ylabel('y values of the array')
plt.title('Scatter Plot of the Two Clusters in our Array')
plt.show()
```

Here we can see the red & blue represent the two different clusters. This is in fact what we expected when we saw the plot from earlier. 

#### Section 3: Algorithm with Three Clusters

Now let's execute the K-means algorithm again with three clusters to see how it differs. 

```python
kmeans = KMeans(n_clusters=3)
kmeans.fit(X)
```

```python
print(kmeans.cluster_centers_)
```

```python
print(kmeans.labels_)
```

```python
plt.scatter(X[:,0], X[:,1], c=kmeans.labels_, cmap='rainbow')
plt.style.use('ggplot')
plt.xlabel('x values of the array')
plt.ylabel('y values of the array')
plt.title('Scatter Plot of Three Clusters in our Array')
plt.show()
```

The three clusters are plotted together in blue, red & yellow. Now let's plot the centroids to see how the points are clustered better.

```python
plt.scatter(X[:,0], X[:,1], c = kmeans.labels_, cmap='rainbow')
plt.style.use('ggplot')
plt.scatter(kmeans.cluster_centers_[:,0], kmeans.cluster_centers_[:,0], color='black')
plt.xlabel('x values of the array')
plt.ylabel('y values of the array')
plt.title('Scatter Plot of the x vs y Values in our Array Including the Centroids')
plt.show()
```

Here are the centroids are black. We can see that in the lower left cluster, there is a very tight cluster. In the remaining two clusters, the points are more scattered.
